````python
# Using ChatGPT to help add a trendline to the plot

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np

fig, ax = plt.subplots(figsize=(5,3))

boulder_monthly_max_clean = boulder_monthly_max.dropna()

x_values = mdates.date2num(boulder_monthly_max_clean.index)


ax.scatter(x_values,
           boulder_monthly_max_clean["TCel"],
           label="Temperature Data",
           color='white',
           edgecolor='black')

# Fit a linear regression model
slope, intercept = np.polyfit(x_values, boulder_monthly_max_clean["TCel"], 1)
trendline = slope * x_values + intercept

ax.plot(x_values, trendline, color='red', label="Trendline")
ax.plot(x_values, boulder_monthly_max_clean["TCel"], color='grey', markersize=0.3)

# Formatting x-axis as dates
# Set x-axis ticks manually to start from 1923
start_date = pd.to_datetime('1923-01-01')
end_date = pd.to_datetime('2023-12-31')
ax.set_xticks(pd.date_range(start_date, end_date, freq='25Y'))
#ax.xaxis.set_major_locator(mdates.YearLocator(5))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# Add text box with slope value (From ChatGPT)
slope_text = f"Slope: {slope:.5f} deg/yr"
ax.text(0.2, 0.1, slope_text, transform=ax.transAxes, fontsize=10,
        verticalalignment='top', bbox=dict(boxstyle='round', facecolor='whitesmoke', alpha=0.9))



ax.set(title="Mean Annual Temperature\nBoulder, CO (1923-2023)",
       ylabel="Temperature (C)")

ax.legend()  # Add legend to display labels

plt.show()
````

````python
# Create interactive map using folium
import folium

# Weather station info
station_lat = 39.99282
station_long = -105.26683
station_location_name = 'NOAA Weather Station [GHCND:USC00050848] - Boulder, CO'

# Create a Folium map object
m = folium.Map(location=[station_lat,station_long], zoom_start=10)

# Add marker for river gage
folium.Marker(
    location=[station_lat, station_long],
    popup=station_location_name,
    icon=folium.Icon(color='black')
).add_to(m)


# Save the map as an HTML file or display it inline
m.save('map.html')
# OR
m
````
